**Exploratory Data Analysis (EDA)**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#dataset loading
df = pd.read_csv("/content/Telco_Customer_Churn_Dataset  (3).csv")
#dataset cleaning
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
df.dropna(inplace=True)
#converting the churn into numeric values
df['Churn']=df['Churn'].map({'Yes':1,'No':0})
sns.set_style("whitegrid")

```

1.Overall Churn Rate

```python
churn_rate = df['Churn'].mean()*100
print(f"Overall Churn Rate: {churn_rate:.2f}%")
plt.figure(figsize=(6,6))
df['Churn'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    labels=['No','Yes'],
    colors=['skyblue','salmon']
)
plt.title("Overall Customer Churn Rate")
plt.ylabel("")
plt.show()
```

2. Gender Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='gender', hue='gender', palette='Set2', legend=False)
plt.title("Customer Distribution by Gender")
plt.show()
```

3. Partner Status

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Partner', hue='Partner', palette='Set1', legend=False)
plt.title("Customer Distribution by Partner Status")
plt.show()
```

4. Dependents

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Dependents', hue='Dependents', palette='Set3', legend=False)
plt.title("Customer Distribution by Dependents")
plt.show()
```

5. Tenure Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['tenure'], bins=30, kde=True)
plt.title("Tenure Distribution")
plt.xlabel("Tenure (Months)")
plt.show()
```

6. Tenure vs Churn

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Churn', y='tenure')
plt.xticks([0,1],["No","Yes"])
plt.title("Tenure vs Churn")
plt.show()
```

7. Churn by Contract Type

```python
contract_churn = pd.crosstab(
    df['Contract'],
    df['Churn'],
    normalize='index'
)

contract_churn.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5),
    color=['skyblue','salmon']
)

plt.title("Churn by Contract Type")
plt.ylabel("Percentage")
plt.legend(["No Churn","Churn"])
plt.show()
```

8. Churn by Payment Method

```python
payment_churn = pd.crosstab(
    df['PaymentMethod'],
    df['Churn'],
    normalize='index'
)

payment_churn.plot(
    kind='bar',
    stacked=True,
    figsize=(10,5),
    color=['skyblue','salmon']
)

plt.title("Churn by Payment Method")
plt.ylabel("Percentage")
plt.legend(["No Churn","Churn"])
plt.xticks(rotation=20)
plt.show()
```
